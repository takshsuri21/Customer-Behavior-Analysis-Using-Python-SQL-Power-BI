### 3. Data Cleaning & Feature Engineering

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median())) 
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
```

```python
df.columns = df.columns.str.replace(' ','_')
```

```python
df  = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#create a column age_group
```

```python
labels = ['Young Adult','Adult','Middle-aged','Senior']
df['age_group'] = pd.qcut(df['age'],q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
#create column purchase_frequency_days

frequency_mapping = {
    'Fortnightly':14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Annually' : 365,
    'Every 3 Months' : 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
df = df.drop('promo_code_used', axis=1)
```

### 4. Loading Cleaned Data into SQL Server


```python
from sqlalchemy import create_engine
import urllib

# Step 1: Connect to SQL Server
server = r"LAPTOP-L2EQTCDF\SQLEXPRESS"   # your SSMS server name
database = "customer_behavior"          # your database name
driver = "ODBC Driver 18 for SQL Server"

connection_string = f"DRIVER={driver};SERVER={server};DATABASE={database};Trusted_Connection=yes;Encrypt=no;"
connection_uri = "mssql+pyodbc:///?odbc_connect={}".format(urllib.parse.quote_plus(connection_string))

engine = create_engine(connection_uri)

# Step 2: Load DataFrame into SQL Server
table_name = "customer"

df.to_sql(table_name, engine, if_exists='replace', index=False)

print(f"Data successfully loaded into table '{table_name}' in database '{database}'.")
```
